```python
#MANAS TIWARI 2023UG1031

#importing numpy and pandas
import pandas as pd
import numpy as np

#reading AQI_data.csv
df = pd.read_csv('AQI_Data.csv')
data = np.array(df)

# (a) - Display the first 5 rows
print('First 5 rows of the dataset :-')
print(df.head(5))

#(b) - Display the last 6 rows
print('\nLast 6 rows the dataset :-')
print(df.tail(6))

#(c) - Show the summary statistics of the dataset
print('\nSummary of the dataset :-')
print(df.describe())

#(d) - Compute the mean AQI of each city of the dataset
AQI = df.groupby('City')['AQI'].apply(lambda x: np.mean(x.to_numpy()))
PM10 = df.groupby('City')['PM10'].apply(lambda x: np.mean(x.to_numpy()))
PM25 = df.groupby('City')['PM2.5'].apply(lambda x: np.mean(x.to_numpy()))

print('\nMean AQI of each city :-')
print(AQI)
print('\nMean PM2.5 of each city :-')
print(PM25)
print('\nMean PM10 of each city :-')
print(PM10)

#Set - 3
PM25_morethen_100 = df['PM2.5']>100
row_list = df[PM25_morethen_100]  # filtered data
print("\nRows having PM2.5 more than 100 :-")
print(row_list)

row_count = {}      # dictionary to store count of each city
for i in row_list['City']:
    if i in row_count:
        row_count[i] += 1
    else:
        row_count[i] = 1   
        
print("\nCount of rows having PM2.5 more than 100 in each city :-")
for i,j in  row_count.items():
    print(f"{i} = {j}")
    
```
